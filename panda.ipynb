```python
#Group by
#Create DataFrames
data={'Brand':['NIKE','USPA','WOODLAND','UCB','PUMA','UCB'],
      'Person':['James','Amy','Charlie','Sarah','John','Jim'],
      'Sales':[200,250,180,150,350,248]      
      }
df=pd.DataFrame(data)
df
```

```python
df.groupby('Brand')
```

```python
by_brand=df.groupby('Brand')
```

```python
by_brand.mean() #takes mean of two same brand
```

```python
#more examples of aggregate methods
by_brand.std()
```
